```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("/Users/Downloads/insurance.csv")
df.head()
```

```python
plt.scatter(df["age"], df["charges"])
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Relationship between Age and Insurance Charges")
plt.show()
```

```python
# Create a model object
model = LinearRegression()

# Fit the model to the data
model.fit(df[["age"]], df["charges"])
```

```python


# Generate some sample ages
ages = [20, 30, 40, 50]

# Predict charges for the sample ages
predicted_charges = model.predict([[age] for age in ages])
```

```python
plt.scatter(df["age"], df["charges"])
plt.plot(ages, predicted_charges, color="red")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.title("Relationship between Age and Insurance Charges with Best-Fit Line")
plt.show()
```
